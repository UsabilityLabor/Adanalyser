```python
signal_array = signal_df.to_numpy()
print("l채nge signal_array=",len(signal_array))
```

Visualize the signal

```python
plt.figure(figsize=(15,10))
plt.plot(signal_array)
plt.title('raw signal')
plt.ylabel('Intensity')
plt.xlabel('data point i')
plt.show()
```

```python
def detrend_signal(signal_s):    
    return signal.detrend(signal_s)
```

```python
signal_array_detrended = signal.detrend(signal_array,type='linear')#detrend_signal(signal_array)
```

Visualize the detrended signal

```python
plt.figure(figsize=(15,10))
#plt.plot(signal_array,label="raw data")
plt.plot(signal_array_detrended, label="detrended data")
plt.ylabel('Intensity')
plt.xlabel('data point i')
plt.legend(loc='best')
plt.show()
```
